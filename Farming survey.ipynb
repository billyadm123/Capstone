```python
#Hhcode is just the household identification number and should'nt have any affect on the predictions
crops = df[['cadm0','hhcode','s1p1c1', 's1p1c2', 's1p1c3','s1p1c4','s1p1c5',
           's2p2c1','s2p2c2','s2p2c3','s2p2c3','s2p2c4','s2p2c5']]
```

# Data cleaning

```python
#I used the 50th percentile to determine the value to use to fill the null values(there were a lot!)
#I decided to use 26.0 to fill all nulls
crops.describe()
```

```python
crops.fillna(26.0, inplace=True)
```

```python
crops.isnull().sum()
```

```python
X = crops.drop('cadm0', axis=1)
y = crops.cadm0
```

# Can the crops grown from this survey be used to predict what crops should be grown in what country in the future and be used as a model for other locations all over the world suffering from food insecurity?

# Linear regression

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30, random_state=42)
```

```python
#Scaling data to improve computing time
from sklearn.preprocessing import StandardScaler
ss = StandardScaler(with_mean=True)
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

```python
#fitting the train test split to linear regression
#L1 and L2 regularization had no affect on the results
#Using Stochastic Gradient Decsent improved to .077 from .081 but increased the mean squared error increased
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred =lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
print(sklearn.metrics.r2_score(y_test, pred))
print(mean_squared_error(y_test, pred))
```

```python
#Visualizing the accuracy of the model compared to the real data, the red line is the predicted and the yellow is the actual
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(y_test, pred, edgecolors=(0, 0, 0))
ax.plot(y_test, y_test, 'k--', lw=4, color='y')
sns.regplot(x=y_test, y=pred, ax=ax, line_kws={"color": "red"})
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')

ax.legend()
```

# Logistic regression

```python
#'multinomial loss' - predicts probabilities of different possible outcomes - accuracy went up by 20%
#'saga' is used for larger datasets and can handle no penalty and multinominal loss(cross-entropy)
#Cross-entropy increases as the predicted probability moves away from actual label
#Cross-entropy calculates total entropy between the probability distributions(difference between the two distributions)
#With Lasso regularization, accuracy went up one percent
from sklearn.linear_model import LogisticRegression
logR = LogisticRegression(penalty='l1', random_state=0, solver='saga', multi_class='multinomial')
```

```python
logR.fit(X_train,y_train)
```

```python
predictions = logR.predict(X_test)
```

```python
#I'm assuming that the orders of the countries are the same as the coded countries
#There is different amounts of data available for each country which is our target variable
df['adm0'].value_counts(normalize=True)
```

```python
#Logistic regression did better with 82% accuracy
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

# K nearest neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=1)
clf.fit(X_train, y_train)
```

```python
test_preds = clf.predict(X_test)
```

```python
#KNN did better than both models with 92% accuracy
from sklearn.metrics import classification_report
print(classification_report(y_test, test_preds))
```
